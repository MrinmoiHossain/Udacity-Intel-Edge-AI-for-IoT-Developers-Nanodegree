### Graffiti in Buttons

Sometimes you'll find Graffiti that are triggered by buttons. For example, here's a button that starts a Graffiti that draws a picture.

Graffiti can also be combined with other Notebook functionality. Sometimes they are hidden away in a button. Such buttons can run code in a terminal shell right in your notebook. 

Clicking the button does two things in succession: it saves a cell's contents into a file, and runs a command in the terminal. For instance, in the cell below there's some python code. Clicking the Run Code button (or pressing Ctrl-enter after clicking inside the code cell) will invoke the python interpreter in the shell to run the code in that cell, after Graffiti saves it for you in _example1.py_. (You cannot specify the filename, the instructor does that for you).

```python
def say_hi(name):
    print('Greetings, most honorable', name, '...and how are you doing today?')

say_hi('Will')
```
